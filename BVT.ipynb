```python
from google.colab import drive
drive.mount('/content/drive')

import sys
import os
prefix = '/content/drive/My Drive/Colab Notebooks/'
sec_w_price_path = prefix + "sec_w_price/"
sys.path.append(sec_w_price_path)
# print(sys.path)

def predict_next(csv, cur_price, last_price):
  data_path = os.path.join(sec_w_price_path, csv)
  print('Path to data: {}'.format(data_path))

  import numpy as np
  import pandas as pd
  pd.options.mode.copy_on_write = True

  df = pd.read_csv(data_path)
  df_10K = df.loc[df["form"] == "10-K"]
  df_10K.index = range(len(df_10K))
  price = df_10K["PriceVar60"]
  df_10K = df_10K.drop(columns=["filed", "form", "Name", "Ticker", "CIK", "Currency", "PriceVar5", "PriceVar10", "PriceVar20", "PriceVar30", "PriceVar60"])
  df_10K.fillna(0, inplace=True)
  price.fillna(0, inplace=True)
  print(df_10K.end.iloc[-1])

  from sklearn.linear_model import LinearRegression

  quant = pd.DataFrame(df_10K)
  quant.end = quant.end.apply(lambda x: pd.to_datetime(x).toordinal())

  model = LinearRegression()

  X = quant
  y = price

  model.fit(X, y)

  predicted = model.predict(np.array(quant.iloc[-1,:]).reshape(1,-1))

  import matplotlib.pyplot as plt
  import matplotlib.dates
  import datetime

  plt.gca().xaxis.set_major_formatter(matplotlib.dates.DateFormatter("%Y-%m-%d"))
  dates = [datetime.datetime.strptime(str(date), "%Y-%m-%d").date() for date in df_10K["end"].values]
  future_dates = [datetime.date(y, 9, 28) for y in range(2023, 2033)]
  plt.gcf().autofmt_xdate()
  plt.close()

  price = pd.concat([price, pd.Series((cur_price - last_price) / (last_price))])

  plt.plot((dates + [datetime.date(2022, 9, 28)]), price, label="Observed")
  plt.plot((dates[-1], datetime.date(2022, 9, 28)), (price.iloc[-2], predicted[0]), label="Prediction", color="red")
  plt.legend()
  plt.xlabel("Date")
  plt.ylabel("Price Change")
  plt.title("Price Change Observed vs. Forecasted")
  plt.show()

  from statsmodels.tsa.arima.model import ARIMA

  price = price[:-1]
  model = ARIMA(price, order=(3, 0, 2))
  model_fit = model.fit()

  forecast = model_fit.forecast(steps=10)

  plt.plot(dates, price, label="Observed")
  plt.plot(future_dates, forecast, label="Prediction", color="red")
  plt.xlabel("Date")
  plt.ylabel("Price Change")
  plt.title("Price Change Observed vs. Forecasted")
  plt.legend()
  plt.show()
```

```python
from tensorflow import keras
from keras import optimizers
from keras.datasets import cifar10
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential, load_model
from keras.layers import Dense, Dropout, Activation, Flatten, LSTM
from keras.layers import Conv2D, MaxPooling2D
from keras.callbacks import ModelCheckpoint
import math
from keras.utils import to_categorical
```

```python
batch_size = 4
epochs = 25
```

```python
def split_data(data, n):
  X, y = [], []
  for i in range(len(data) - n + 1):
    X.append(data.iloc[i:i + n, :-1])
    y.append(data.iloc[i + n - 1, -1])
  return np.array(X), np.array(y)

aapl_data = pd.DataFrame(aapl_10K)
# aapl_data["price"] = pd.Series(aapl_price)
aaplX, aaply = split_data(aapl_data, 2)
aaplX.shape, aaply.shape
```

```python
x_train, x_test = aapl_10K[:12], aapl_10K[12:]
y_train, y_test = aapl_price[:12], aapl_price[12:]
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train, x_test = aaplX[:12], aaplX[12:]
y_train, y_test = aaply[:12], aaply[12:]
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
lstm = Sequential()
lstm.add(LSTM(32, input_shape=(x_train.shape[1], x_train.shape[2]), activation="relu", return_sequences=True))
lstm.add(Dense(1))
lstm.compile(loss="mean_squared_error", optimizer="adam")
lstm.summary()
```

```python
history = lstm.fit(x_train, y_train, epochs=epochs, batch_size=batch_size, verbose=2, shuffle=False)
```

```python
pred = lstm.predict(x_test)
```
